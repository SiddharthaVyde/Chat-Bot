```python
import pickle
import numpy as np
```

```python
with open('train_qa.txt','rb') as f:
    train_data = pickle.load(f)
```

```python
with open('test_qa.txt','rb') as f:
    test_data = pickle.load(f)
```

```python
data = train_data + test_data
```

```python
vocab = set()

for story,question,answer in data:
    vocab = vocab.union(set(story))
    vocab = vocab.union(set(question))
```

```python
vocab.add('no')
```

```python
vocab.add('yes')
```

```python
vocab
```

```python
vocab_len = len(vocab) + 1
```

```python
all_story_len = [len(d[0]) for d in data]
all_story_len
```

```python
max_story_len = max(all_story_len)
```

```python
max_ques_len = max([len(d[1]) for d in data])
```

```python
from keras_preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(filters=[])
tokenizer.fit_on_texts(vocab)
```

```python
tokenizer.word_index
```

```python
train_story_text = []
train_questions_text = []
train_answers = []
```

```python
for story,question,answer in train_data:
    train_story_text.append(story)
    train_questions_text.append(question)
    train_answers.append(answer)
```

```python
train_story_text
```

```python
train_story_seq = tokenizer.texts_to_sequences(train_story_text)
```

```python
train_story_seq
```

```python
def vectorize_stories(data,word_index=tokenizer.word_index,max_story_len=max_story_len,max_ques_len=max_ques_len):
    X = [] #stories = X
    Xq = [] #Questions
    Y = [] #Correct answer
    
    for story,ques,ans in data:
        x = [word_index[word.lower()] for word in story]
        xq = [word_index[word.lower()] for word in ques]
        y = np.zeros(len(word_index)+1)
        y[word_index[ans]] = 1
        
        X.append(x)
        Xq.append(xq)
        Y.append(y)
        
    return(pad_sequences(X,maxlen=max_story_len),pad_sequences(Xq,maxlen=max_ques_len),np.array(Y))
```

```python
input_train,ques_train,ans_train = vectorize_stories(train_data)
```

```python
input_test,ques_test,ans_test = vectorize_stories(test_data)
```

```python
input_test
```

```python
ans_test
```

```python
sum(ans_test)
```

```python
from keras.models import Sequential, Model
from keras.layers import Input, Activation, Dense, Permute, Dropout,add, dot, concatenate, LSTM, Embedding
```

```python
input_seq = Input((max_story_len,))
ques = Input((max_ques_len,))
```

```python
vocab_size = len(vocab) + 1
```

```python
#Input encoder M
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim=vocab_size,output_dim=64))
input_encoder_m.add(Dropout(0.3))
# gives (samples,story_maxlen,embedding_dim)
```

```python
#Input encoder C
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim=vocab_size,output_dim=max_ques_len))
input_encoder_c.add(Dropout(0.3))
# gives (samples,story_maxlen,max_ques_len)
```

```python
ques_encoder = Sequential()
ques_encoder.add(Embedding(input_dim=vocab_size,output_dim=64,input_length=max_ques_len))
ques_encoder.add(Dropout(0.3))
# gives (samples,max_ques_len,embedding_dim)
```

```python
# encoded = encoder(input)
input_encoded_m = input_encoder_m(input_seq)
input_encoded_c = input_encoder_c(input_seq)
ques_encoded = ques_encoder(ques)
```

```python
match = dot([input_encoded_m,ques_encoded],axes=(2,2))
match = Activation('softmax')(match)
```

```python
response = add([match,input_encoded_c])
response = Permute((2,1))(response)
```

```python
answer = concatenate([response,ques_encoded])
answer
```

```python
answer = LSTM(32)(answer)
answer = Dropout(0.5)(answer)
answer = Dense(vocab_size)(answer)
# gives (samples,vocab_size) YES/NO 0000 
```

```python
answer = Activation('softmax')(answer)
```

```python
model = Model([input_seq,ques],answer)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit([input_train,ques_train],ans_train,batch_size=32,epochs=100,validation_data=([input_test,ques_test],ans_test))
```

```python
filename = 'chatbot_120_epochs.h5'
model.save(filename)
```
